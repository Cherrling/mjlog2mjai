```python

# 读取links文件，去掉其中不是链接的元素，最后输出到一个新文件中

import re
with open('links', 'r') as f:
    lines = f.readlines()
    
with open('links_new', 'w') as f:
    for line in lines:
        if re.match(r'^http', line):
            f.write(line)
        else:
            print('not a link:', line)
```

```python
# 读取ori_gz目录下的全部文件，将其中的所有链接提取出来，输出到paipu_links文件中
import os
import re
# 使用utf-8编码打开文件
def open_file(file):
    with open(file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    return lines
with open('paipu_links', 'w') as f:
    for root, dirs, files in os.walk('ori_gz'):
        for file in files:
            lines = open_file(os.path.join(root, file))
            for line in lines:
                # 00:00 | 14 | 四鳳東喰赤速 | <a href="http://tenhou.net/0/?log=2024092500gm-00e1-0000-c2276218">牌譜</a> | 桜花閃々(+45.2) 朝花夕歌(+5.5) 俺と妖夢(-19.5) つっぱりポール(-31.2)<br>
                # 每行类似于上面的格式，提取出其中的链接
                match = re.search(r'href="(.+?)"', line)
                print(match)
                if match:
                    f.write(match.group(1) + '\n')
                else:
                    print('not a link:', line)
```
